```python
# 5:
rgn120 = reign_csv[(reign_csv['tenure_months'] < 120)]
dfgb = pd.DataFrame(rgn120.groupby(['leader', 'country', 'male']).agg({'age': 'max'}).reset_index())
print("Среднее и среднеквадратичное отклонение возрастов правителей со сроком < 120 мес.: {0:.3f} {1:.3f}"
      .format(dfgb.age.mean(), dfgb.age.std()))
rgn120 = reign_csv[(reign_csv['tenure_months'] >= 120)]
dfgb = pd.DataFrame(rgn120.groupby(['leader', 'country', 'male']).agg({'age': 'max'}).reset_index())
print("Среднее и среднеквадратичное отклонение возрастов правителей со сроком >= 120 мес.: {0:.3f} {1:.3f}"
      .format(dfgb.age.mean(), dfgb.age.std()))
```
